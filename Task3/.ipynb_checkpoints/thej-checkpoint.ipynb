```python
# 3.2
import math 
from scipy.constants import G

def v_2(m, r):
  vel = math.sqrt((2*G*m)/r)
  return vel

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

for m, r, v in zip(masses, radii, names):
    print(f" V_2 = {v_2(m, r*1000)},  Mass = {m:.2f}. Name: {v}.")

planetaryslop = dict()
for m, r, v in zip(masses, radii, names): 
    planetaryslop.update({v: v_2(m, r*1000)})
print(planetaryslop)
    


```

```python
# 3.10
import math
def cylinder(radius, height):
    area = 2*math.pi*(radius**2)+2*math.pi*radius*height
    volume = math.pi*(radius**2)*height
    return area, volume
    
a, v = cylinder(6, 7)
ratio = a/v

print(ratio)
```

```python
# 3.17
import numpy as np 
import math

def mean(array):
    sum = 0
    for i in array:
        sum += i
    return sum/len(array)

def stdev(array):
    mid = 0
    mean2 = mean(array)
    for j in array:
        mid += (j - mean2)**2
    return math.sqrt(mid/len(array))

myarray = list(np.random.normal(2, 3, 100))

print(mean(myarray))
print(stdev(myarray))
print(np.mean(myarray))
print(np.std(myarray))

sigma = np.sqrt((1/len(myarray))*((np.sum((myarray-np.mean(myarray))**2))))
print(sigma)
```

```python
# 3.22
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
fig, ax = plt.subplots()

x = np.arange(-5, 6)
y = np.cosh(x)
plt.scatter(x, y)
plt.plot(x, y)
ax.set_title("Catenary plot", 
              
ax.set_ylabel("Height (m)", fontsize=20)
ax.set_xlabel("Length (m)", fontsize =20)
red_patch = patches.Patch(color = "blue", label = "Catenary")
ax.legend(handles=[red_patch])
ax.grid()
plt.show()
```
