```python
# 6.1

import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from scipy import stats as st

fig, ax = plt.subplots(2)

hdul = fits.open('m101.fits')
dat = hdul[0].data
juan = hdul[1].header
dat = dat.flatten()

npmin = np.min(dat)
npmax = np.max(dat)
npstd = np.std(dat)
npmean = np.mean(dat)
npmedian = np.median(dat)
npmode, _ = st.mode(dat)
print(npmin)
print(npmax)
print(npstd)
print(npmean)
print(npmedian)
print(npmode)

arraydata = np.reshape(np.asarray(dat), (530, 530))

ax[0].hist(dat, bins=500)
ax[0].axvline(npmean, color='r', label = 'mean')
ax[0].axvline(npmedian, color='violet', label = 'median')
ax[0].axvline(npmode, color='b', label = 'mode')
ax[0].legend()

shower = ax[1].imshow(arraydata, cmap = 'YlOrBr')
cbar = fig.colorbar(shower)
cbar.set_label("COUNT")
ax[1].set_xlabel("CTYPE1")
ax[1].set_ylabel("CTYPE2")

fig.set_figheight(10)
fig.set_figwidth(10)

plt.show()

print(np.shape(dat))

juan['BUNIT'] = 'Counts'

```

```python
# 6.5

from astropy.io import fits
import numpy as np

hdul0 = fits.getdata('noisemap0.fits')
hdul1 = fits.getdata('noisemap1.fits')
hdul2 = fits.getdata('noisemap2.fits')

sigma0 = np.std(hdul0)
sigma1 = np.std(hdul1)
sigma2 = np.std(hdul2)

print(sigma0)
print(sigma1)
print(sigma2)

stack = np.stack((sigma0, sigma1, sigma2), axis=0)
mean = np.mean(stack, axis=0)
sigma_mean = np.std(mean)

print(sigma_mean)

N = 3
sigma_again = (1/N)*np.sqrt(sigma0**2+sigma1**2+sigma2**2)

print(sigma_again)

# they aren't comparable
```

```python
# 6.3

import numpy as np
from astropy.io import fits

data = fits.getdata('m101.fits')

slicer = data[50:70, 50:70].astype(np.float64)

pixel = slicer.flatten()

mean = np.mean(pixel)
sigma = np.std(pixel)
rms = np.sqrt(np.mean(pixel**2))

print(mean)
print(sigma)
print(rms)

# DIffers because std measures data spread rel to mea n while rms has no mention of it so they are the same when mean is 0 
```

```python
# 6.10

import sqlite3 as sqrlite
import platform
from astroquery.gaia import Gaia
from astropy.table import QTable
from astropy.io import ascii
import pandas as pd

simple_query = """
    SELECT TOP 50
    source_id,
    ra, dec,
    parallax,
    parallax_error,
    phot_g_mean_mag,
    phot_bp_mean_mag,
    phot_rp_mean_mag
FROM gaiadr3.gaia_source
WHERE parallax/parallax_error>5
    AND phot_g_mean_mag IS NOT NULL
    AND phot_bp_mean_mag IS NOT NULL
    AND phot_rp_mean_mag IS NOT NULL
ORDER BY RANDOM_INDEX
"""

print("Sending query to Gaia archive...")
job = Gaia.launch_job(simple_query)
results = job.get_results()
gaia = results.to_pandas()
data = QTable.from_df(gaia)
ascii.write(data, 'astroquery_gaia.fits', overwrite = True)
```

```python
# 6.14

class Fibonacci: 
    """Calculates Fibonacci sequence"""
    
    def get_nth_term(self, n):
           
        a, b = 0, 1
        for _ in range(2, n+1):
            a, b = b, a + b
        return b
    def get_divisible(self, n, m):
        
        divisible = []
        a, b = 0, 1
        
        for i in range(n):
            if a % m == 0:
                divisible.append(a)
            a, b = b, a+b
            
        return divisible 
    
n = 100
m = 7

fib = Fibonacci()

nth = fib.get_nth_term(n)
divlist = fib.get_divisible(n, m)

print(nth)
print(divlist)
```
